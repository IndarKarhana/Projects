


# EDA

As we can see in our data that the dtype of Date_of_Journey  column is "object", we need to change it into "datetime" dtype so that we can work with that data. We use pandas "to_datetime" function to convert the dtype.

.dt.day method will extract only the day of that date
.dt.month method will extract only the month of that date





```python
training_data["Journey_day"] = pd.to_datetime(training_data.Date_of_Journey, format="%d/%m/%Y").dt.day
```

```python
training_data["Journey_month"] = pd.to_datetime(training_data["Date_of_Journey"], format="%d/%m/%Y").dt.month
```

```python
training_data.head()
```

```python
#we drop our Date_of_Journey column as we have already extractd useful info from it
training_data.drop(["Date_of_Journey"],axis = 1, inplace = True)
```

```python
#departure time is the time when plane leaves the gate so similar to Date_of_Journey we can extract values from it in datetime dtype

#Extracting hours
training_data["Dep_hour"] = pd.to_datetime(training_data["Dep_Time"]).dt.hour

#Extracting Minutes
training_data["Dep_min"] = pd.to_datetime(training_data.Dep_Time).dt.minute

#Now we can drop Dep_Time as it is of no use
training_data.drop(["Dep_Time"], axis = 1, inplace = True)
```

```python
training_data.head()
```

```python
#Arrival time is when the plane pulls up to the gate.
#Similar to Date_of_Journey we can extract valus from Arrival_Time

#Extracting hours
training_data["Arrival_hour"] = pd.to_datetime(training_data["Arrival_Time"]).dt.hour

#Extracting Minutes
training_data["Arrival_min"] = pd.to_datetime(training_data.Arrival_Time).dt.minute

#Now we can drop Arrival_Time as it is of no use
training_data.drop(["Arrival_Time"], axis = 1, inplace = True)
```

```python
training_data.head()
```

```python
# time taken by plane to reach destination is called duration, also its difference b/w dep_time and arrival_time
#now we will work on column Duration and extract workable data from it.


duration = list(training_data["Duration"])

for i in range(len(duration)):
    if len(duration[i].split()) != 2:
        if "h" in duration[i]:
            duration[i] = duration[i].strip() + " 0m"
        else:
            duration[i] =  "0h " + duration[i]
duration_hours = []
duration_minutes = []
for i in range(len(duration)):
    duration_hours.append(int(duration[i].split(sep  = "h")[0]))   #extract hours from duration list
    duration_minutes.append(int(duration[i].split(sep = "m")[0].split()[-1])) #extract minutes from duration list
```

```python
#Adding Duration_hours and Duration_mins columns to the training_data
training_data["Duration_hours"] = duration_hours
training_data["Duration_mins"] = duration_minutes
```

```python
#Dropping the Duration column
training_data.drop(["Duration"], axis = 1, inplace = True)
```

```python
training_data.head()
```

# Handling Categorical Data

Some general ways to handle categorical data are:

1. Nominal Data : data are not in any order > OneHotEncoder is used in this case
2. Ordinal Data : data are in order > LabelEncoder is used in this case

```python
training_data["Airline"].value_counts()
```

```python
#Airline vs Price
sns.catplot(y = "Price", x = "Airline", data = training_data.sort_values("Price",ascending = False), kind = "boxen", height = 6, aspect = 4)
plt.show()
```

```python
#from graph we can see that jet airways buisness have the heighest price.
#apart from the first airline almost all are having similar



# As airlines is a Nominal Categorical data we will perform OneHotEncoding

Airline = training_data[["Airline"]]

Airline = pd.get_dummies(Airline, drop_first = True)    # drop_first = True will drop the first column made after OneHotEncoding coz it can be easily specified using the other columns

Airline.head()
```

```python
#now we look for the unique values for the source column
training_data["Source"].value_counts()
```

```python
#source vs price
sns.catplot(y = "Price", x = "Source", data = training_data.sort_values("Price",ascending = False), kind = "boxen", height = 6, aspect = 4)
```

```python
# As Source is Nominal Categorical data we will perform OneHotEncoding

Source = training_data[["Source"]]

Source = pd.get_dummies(Source, drop_first = True)

Source.head()
```

```python
#now we look for the unique values for the Destination column
training_data["Destination"].value_counts()
```

```python
# As Destination is Nominal Categorical data we will perform OneHotEncoding

Destination = training_data[["Destination"]]
Destination = pd.get_dummies(Destination, drop_first = True)
Destination.head()
```

```python
training_data["Route"]
```

```python
# Additional_Info column contains 80% no-info
# Route and Total_Stops are related to each other

#so we can drop both additional_info and route column
training_data.drop(["Route", "Additional_Info"], axis = 1, inplace = True)
training_data.head()
```

```python
training_data["Total_Stops"].value_counts()
```

```python
# As this is the case of Ordinal Categorical type we perform LabelEncoder
# Here Values are assigned with corresponding keys

training_data.replace({"non-stop": 0, "1 stop": 1, "2 stops": 2, "3 stops": 3, "4 stops": 4}, inplace = True)
```

```python
training_data.head()
```

```python
# now we concatenate all the information we have gathered so far by working on raw data

data_train = pd.concat([training_data, Airline, Source, Destination], axis = 1)
```

## Hyperparameter Tuning

```python
#we will use RandomisedSearchCV as it is faster 
from sklearn.model_selection import RandomizedSearchCV
```

```python
#no. of trees in random forest
n_estimators = [int(x) for x in np.linspace(start = 100, stop = 1200, num =12)] #linspace gives us equally spaced numbers based on intervals

#no. of features to consider every split
max_features = ['auto', 'sqrt']

#max no. of levels in tree
max_depth = [int(x) for x in np.linspace(5, 30, num = 6)]

#min no. of samples required to split a node
min_samples_split = [2, 5, 10, 15, 100]

#min no.of samples required at each leaf node
min_samples_leaf = [1, 2, 5, 10]
```

```python
#create random grid

random_grid = {'n_estimators': n_estimators,
               'max_features': max_features,
               'max_depth': max_depth,
               'min_samples_split': min_samples_split,
               'min_samples_leaf': min_samples_leaf}
```

```python
# random search using 5 fold cv 
rf_random = RandomizedSearchCV(estimator = reg_rf, param_distributions = random_grid, scoring = 'neg_mean_squared_error',
                               n_iter =10, cv = 5, verbose = 2, random_state = 42, n_jobs = 1)
```

```python
rf_random.fit(X_train,y_train)
```

```python
rf_random.best_params_
```

```python
prediction = rf_random.predict(X_test)
```

```python
plt.figure(figsize = (8,8))   #to check our assumption that it follows our absolute error folows gaussian distribution
sns.distplot(y_test-prediction)
plt.show()
```

```python
plt.figure(figsize = (8,8))
plt.scatter(y_test, prediction, alpha = 0.5)
plt.xlabel("y_test")
plt.ylabel("y_pred")
plt.show()
```

```python
print("MAE: ", metrics.mean_absolute_error(y_test, prediction))
print("MSE: ", metrics.mean_squared_error(y_test, prediction))
print("RMSE: ", np.sqrt(metrics.mean_squared_error(y_test, prediction)))
```
